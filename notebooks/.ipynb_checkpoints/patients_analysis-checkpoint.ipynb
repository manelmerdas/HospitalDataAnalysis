```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV
df = pd.read_csv("../data/patients.csv")

# Aperçu des données
print(df.head())

# Statistiques descriptives
print(df.describe())

# Visualisation : Cholestérol vs Âge
plt.plot(df["Age"], df["Cholesterol"], marker="o")
plt.xlabel("Âge")
plt.ylabel("Taux de cholestérol")
plt.title("Relation entre âge et cholestérol")
plt.show()
```
